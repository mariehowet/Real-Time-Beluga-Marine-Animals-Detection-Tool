```python
# python
from pathlib import Path
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Répertoire outputs
outputs = Path(r"C:\UQACode\Automne2025\Real-Time-Beluga-Whales-Detection-Tool\outputs")
if not outputs.exists():
    outputs = Path.cwd()

def find_col_like(df, keywords):
    for col in df.columns:
        lc = col.lower()
        if all(k in lc for k in keywords):
            return col
    # fallback: any column that contains any keyword
    for col in df.columns:
        lc = col.lower()
        if any(k in lc for k in keywords):
            return col
    return None

def parse_to_percent(v):
    if pd.isna(v):
        return np.nan
    s = str(v).strip().replace("%", "")
    try:
        val = float(s)
    except Exception:
        return np.nan
    if 0 <= val <= 1:
        val = val * 100
    return val

rows = []
phase_dirs = sorted([p for p in outputs.glob("Phase *") if p.is_dir()])

for phase in phase_dirs:
    files = list(phase.glob("models_comparison*.csv"))
    if not files:
        continue
    try:
        df = pd.read_csv(files[0])
    except Exception:
        continue

    mcol = find_col_like(df, ["map", "50"])
    if mcol is None:
        mcol = find_col_like(df, ["map"])
    if mcol is None:
        continue
    df["_mAP50_pct"] = df[mcol].apply(parse_to_percent)
    if df["_mAP50_pct"].dropna().empty:
        continue
    best = df.loc[df["_mAP50_pct"].idxmax()]

    # extraire métriques - reconnaitre variantes de noms
    prec_col = find_col_like(df, ["precis", "precision"])
    rec_col  = find_col_like(df, ["recall", "rappel"])
    f1_col   = find_col_like(df, ["f1", "f-score", "fscore"])

    prec = parse_to_percent(best[prec_col]) if prec_col is not None and prec_col in best else np.nan
    rec  = parse_to_percent(best[rec_col])  if rec_col  is not None and rec_col in best  else np.nan
    f1   = parse_to_percent(best[f1_col])   if f1_col   is not None and f1_col in best   else np.nan

    # calculer F1 si manquant et si prec/rec disponibles (en % -> convertir en fraction pour formule)
    if (pd.isna(f1)) and (not pd.isna(prec)) and (not pd.isna(rec)):
        p = prec / 100.0
        r = rec / 100.0
        if (p + r) > 0:
            f1 = 2 * p * r / (p + r) * 100.0

    rows.append({
        "Phase": phase.name.split("-")[0],
        "mAP@50": float(best["_mAP50_pct"]),
        "Précision": float(prec) if not pd.isna(prec) else np.nan,
        "Rappel": float(rec) if not pd.isna(rec) else np.nan,
        "F1": float(f1) if not pd.isna(f1) else np.nan
    })

if not rows:
    print("Aucune donnée trouvée.")
else:
    metrics = pd.DataFrame(rows).set_index("Phase")
    # conserver ordre trouvé ; remplacer valeurs manquantes par nan
    metrics = metrics[["mAP@50", "Précision", "Rappel", "F1"]]


```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd


# Trier par numéro de phase
def phase_number(p):
    try:
        return int(p.split()[1])
    except:
        return 0

metrics_plot = metrics.reindex(sorted(metrics.index, key=phase_number))

# Bar plot horizontal
phases = metrics_plot.index.tolist()
metrics_names = metrics_plot.columns.tolist()
n_phases = len(phases)
n_metrics = len(metrics_names)
y = np.arange(n_phases)
height = 0.15

fig, ax = plt.subplots(figsize=(10, max(3, 0.6 * n_phases)))

cmap = plt.cm.Blues
colors = [cmap((i + 1) / n_metrics) for i in range(n_metrics)]

# Remplir NaN par 0 pour afficher la barre, mais garder valeur originale pour annotation
metrics_plot_filled = metrics_plot.fillna(0)

for i, (metric, color) in enumerate(zip(metrics_names, colors)):
    bars = ax.barh(y + i * height, metrics_plot_filled[metric], height, label=metric, color=color)

ax.set_yticks(y + height * (n_metrics - 1) / 2)
ax.set_yticklabels(phases)
ax.set_xlim(70, 100)
ax.set_xlabel("%")
#ax.set_title("Métriques des meilleurs modèles par phase lors de l'évaluation sur les images de test")

ax.legend(title="Métriques", bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
out_bar = outputs / "metrics_barplot_best_models_horizontal.png"
#plt.savefig(out_bar, dpi=200, bbox_inches='tight')
plt.show()
print(f"Bar plot horizontal sauvegardé dans `{out_bar}`")
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
import seaborn as sns


root = "Phase 5 - SAM2  & ViT/prediction-videos/predicted"

folders = ["beluga", "dolphin", "other"]

def get_true_class(filename):
    """Déduit la classe réelle à partir du nom du fichier."""
    name = filename.lower()
    if "beluga" in name:
        return "beluga"
    if "dolphin" in name:
        return "dolphin"
    return "other"


# Pour stocker les résultats
records = []

# Parcourt chaque dossier (classe prédite)
for predicted_class in folders:
    folder_path = os.path.join(root, predicted_class)

    # Liste des fichiers du dossier
    for filename in os.listdir(folder_path):
        if filename.lower().endswith((".jpg", ".png", ".jpeg")):
            true_class = get_true_class(filename)
            correct = (true_class == predicted_class)

            records.append({
                "filename": filename,
                "predicted": predicted_class,
                "true": true_class,
                "correct": correct
            })


# Convertir en DataFrame pour analyse
df = pd.DataFrame(records)

# Affichage rapide
print(df.head())
print("\nAccuracy globale :", df["correct"].mean())
print("\nConfusion matrix :")
print(df.groupby(["true", "predicted"]).size())

cm = confusion_matrix(df["true"], df["predicted"], labels=folders, normalize="true")


sns.heatmap(cm, annot=True, cmap="Blues", fmt=".2f",
            xticklabels=folders, yticklabels=folders)
plt.title("Confusion Matrix on Test Set")
plt.xlabel('Predicted')
plt.ylabel('True')
out_cm = f"{root}/confusion_matrix_test_phase5.png"

plt.savefig(out_cm, dpi=200, bbox_inches='tight')
plt.show()
print(f"Matrice de confusion sauvegardée dans `{out_cm}`")
```
